# importing Libraries

```python
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

```python
# Import txt file

with open('Key_events_20th_century.txt', 'r', errors='ignore') as file: 
   data = file.read().replace( '\n', ' ')
```

```python
# Sentence tokenization

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent) 
```

```python
# Word tokenization into single words

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# checking their frequency distribution
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words) 
```

```python
dist_words.most_common(10)    #checking for the most common words
```

```python
# Creating a frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words.plot(10,cumulative = False)
plt.show()
```

```python
# Removing the stopwords in words
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_words = []
for word in tokenized_word:
    if word.lower() not in stop_words:
        filtered_words.append(word)
```

```python
# Creating a new FreqDist for filtered–words

Dist_words_filter = FreqDist(filtered_words)
print (Dist_words_filter)
```

```python
# Creating a frequency Distribution Plot

plt.figure(figsize=(8, 3))
Dist_words_filter.plot(10,cumulative = False)
plt.show()
```

# the most common strings in the text sample are mainly punctuation marks and the year "2018".

```python
# Substituting all punctuation marks with a space

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_words))
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Creating a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2) 
```

```python
# Creating a frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(10,cumulative = False)
plt.show()
```

```python
# Creating a tags list using TextBlob.

new_stopwords = ["s", "th"]

filtered = []
for word in tokenized_word_2:
     if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))    # creating a TextBlob object
```

```python
tags_list = text.tags
print(tags_list)                  #creating POS tags with tags function
```

```python
# Create a list of the Top 10 POS tags for words that appear in the article.

df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', "Word type"]
df_t=df_text.groupby('Word type').count().reset_index()
top10=df_t.nlargest(10, 'Words')
```

```python
# creating a bar plot of the Top 10 POS tags for the words that appear in the article

plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
      sns.barplot(x = "Words", y = "Word type",
     saturation = 0.9, data = top10).set_title("Key_events_of_the_20th_century - top 10 word types used")
```

```python
# Creating three bar plots with the top 15 POS labels—one each for nouns

df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNS")]
df.columns = ["Word", "Occurrences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top15 = y.nlargest(15, 'Occurrences') 
```

```python
import pandas as pd
import re

# Example text (replace with your 'data')
text_data = data.lower()

# Full list of countries and aliases
country_list = [
    "united states", "usa", "us", "america",
    "soviet union", "ussr", "russia",
    "united kingdom", "uk", "britain", "england",
    "germany", "france", "japan", "italy", "china", "india", "canada"
]
```

```python
# Turning into dataframe

df_countries = pd.DataFrame(country_list, columns=["Country"])
```

```python
# Creating alias mapping

df_countries["Country_Alias"] = df_countries["Country"].apply(
    lambda x: "United States" if x in ["united states", "usa", "us", "america"] else
              "Soviet Union" if x in ["soviet union", "ussr", "russia"] else
              "United Kingdom" if x in ["united kingdom", "uk", "britain", "england"] else
              x.title()
)
```

```python
# Counting mentions using the alias column
mention_counts = []
for alias in df_countries["Country_Alias"].unique():
    aliases = df_countries[df_countries["Country_Alias"] == alias]["Country"].tolist()
    pattern = r'\b(?:' + '|'.join(map(re.escape, aliases)) + r')\b'
    matches = re.findall(pattern, text_data)
    mention_counts.append((alias, len(matches)))
```

```python
# Final dataframe
df_final = pd.DataFrame(mention_counts, columns=["Country", "Mentions"])
df_final = df_final.sort_values(by="Mentions", ascending=False).reset_index(drop=True)

print(df_final)
```

```python
# Create the DataFrame in a suitable format

df = pd.DataFrame(mention_counts, columns=["Country", "Mentions"])
```

```python
# Creating a plot showing the frequency of the number of times these countries are mentioned in the text

# Plot the country mentions
plt.figure(figsize=(10, 6))
sns.barplot(x="Mentions", y="Country", data=df, palette="Blues_d")

plt.title("Frequency of Country Mentions in Text", fontsize=16)
plt.xlabel("Number of Mentions")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

### Findings from Country Mentions

From the plot, it is clear that "Soviet Union", "Germany", "United States" and "Japan" have the highest number of mentions as compared to other countries. Also, countries like "Canada" and "India" have the least mentions, suggesting that the text focuses more on the primary influencers of the century.
